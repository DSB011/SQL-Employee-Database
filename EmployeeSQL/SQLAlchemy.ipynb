```python
# Dependencies
# Pandas
import pandas as pd
import psycopg2
import sqlalchemy
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
# SQL Alchemy
from sqlalchemy import create_engine
from config import Password
engine = create_engine('postgresql://postgres:{Password}@localhost:5432/EmployeeSQL_db')
connection = engine.connect()
```

```python
employees_info = pd.read_sql("SELECT * FROM Employees", connection)
salary_info = pd.read_sql("SELECT * FROM Salaries", connection)
title_info = pd.read_sql("SELECT * FROM Titles", connection)
```

```python
salary_info
```

```python
# Bonus 1 :  histogram to visualize the most common salary ranges for employees
plt.hist(salary_info['salary'], 20, label= "Salary")
plt.axvline(salary_info['salary'].mean(), color='r', linestyle = 'dashed', label="Salary Mean")
plt.xlabel("Salary")
plt.ylabel("Number of Employees")
plt.legend()
plt.title("Common Salary Ranges for Employees")
plt.savefig("Common Salary Ranges for Employees.png")
```

```python
# Bonus 2 :  bar chart of average salary by title.
query_pandas = pd.read_sql("SELECT * FROM (SELECT * FROM Employees INNER JOIN Titles ON Employees.emp_title_id = Titles.title_id) AS joined INNER JOIN Salaries ON joined.emp_no=Salaries.emp_no", connection)
```

```python
query_pandas
```

```python
avg_salaries_by_title = query_pandas.groupby(["title"]).mean()["salary"]

# Plotting the Bar Chart
titles_col = avg_salaries_by_title.index
avg_salaries_title = avg_salaries_by_title.values

x_axis = np.arange(len(titles_col))
tick_locations = [range for range in x_axis]
plt.xticks(tick_locations,titles_col, rotation=90)
plt.title("Average Salary By Title")
plt.xlabel("Title")
plt.ylabel("Average Salary")
multi_plot = plt.bar(x_axis, avg_salaries_title, color='r', alpha=0.5, align='center')

for i in multi_plot:
    yvalue = i.get_height()
    plt.text(i.get_x(), yvalue + 300, "${:,.2f}K".format(yvalue/1000))
plt.savefig("Average_Salary_Title.png")
plt.show()

```
